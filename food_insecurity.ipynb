```python
#Converting percentages to values
df_2006.lalowi1 = df_2006.lalowi1*df_2006.Pop2010/100
df_2006.laskids1 = df_2006.laskids1*df_2006.Pop2010/100
df_2006.laseniors1 = df_2006.laseniors1*df_2006.Pop2010/100
df_2006.lahunv1 = df_2006.lahunv1*df_2006.Pop2010/100
df_2006.loc[:,['Year', 'State', 'County', 'Urban', 'Pop2010', 'OHU2010', 'lapop1', 'lalowi1', 'laskids1', 'laseniors1', 'lahunv1']]
```

```python
#ceiling the data to make it integer
import math
df_2006.lapop1 = df_2006.lapop1.apply(lambda x: math.ceil(x))
df_2006.lalowi1 = df_2006.lalowi1.apply(lambda x: math.ceil(x))
df_2006.laskids1 = df_2006.laskids1.apply(lambda x: math.ceil(x))
df_2006.laseniors1 = df_2006.laseniors1.apply(lambda x: math.ceil(x))
df_2006.lahunv1 = df_2006.lahunv1.apply(lambda x: math.ceil(x))
df_2006.loc[:,['Year', 'State', 'County', 'Urban', 'Pop2010', 'OHU2010', 'lapop1', 'lalowi1', 'laskids1', 'laseniors1', 'lahunv1']]
```
