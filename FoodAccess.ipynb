```python

import pandas as pd
import numpy as np
from scipy.stats import norm
```

```python
foodAccesss_data_Uncleaned = pd.read_excel('FoodAccessResearchAtlasData2019.xlsx')
foodAccesss_data_Uncleaned.head()
```

```python
columns_to_sum = ['Pop2010',	'OHU2010', 'lapophalf',	'lalowihalf',	'lakidshalf',	'laseniorshalf',	'lawhitehalf',	'lablackhalf',	'laasianhalf',	'lanhopihalf',	'laaianhalf',	'laomultirhalf',	'lahisphalf',	'lahunvhalf',	'lasnaphalf',	'lapop1',	'lalowi1',	'laseniors1',	'lawhite1',	'lablack1',	'laasian1',	'lanhopi1',	'laaian1',	'laomultir1',	'lahisp1',	'lahunv1',	'TractLOWI',	'TractKids',	'TractSeniors',	'TractWhite',	'TractBlack',	'TractAsian',	'TractNHOPI',	'TractAIAN',	'TractOMultir',	'TractHispanic']
columns_to_avg = ['PovertyRate', 'MedianFamilyIncome']
col = ['Year', 'State', 'County']

# Calculate sum of selected columns grouped by 'Group'
sum_by_group = foodAccesss_data_Uncleaned.groupby(['County', 'Urban'])[columns_to_sum].sum().reset_index()
avg_by_group = foodAccesss_data_Uncleaned.groupby(['County', 'Urban'])[columns_to_avg].mean().reset_index()

merged_df = pd.merge(sum_by_group, avg_by_group, on=['County', 'Urban'], how='inner', suffixes=('_df1', '_df2'))

county_contryPairs = foodAccesss_data_Uncleaned[col].drop_duplicates()

merged_df_1 = pd.merge(county_contryPairs[col], merged_df, on = 'County', how = 'inner', suffixes=('_df1', '_df2'))



merged_df_1.to_csv('FoodAccess19output13.csv', index=True)
```
